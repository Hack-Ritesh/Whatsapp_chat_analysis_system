```python
def count_word_occurrences(chat_series, search_word):
    # Combine all messages into a single string
    all_messages = ' '.join(chat_series)

    # Counting occurrences of the search word
    word_count = Counter(re.findall(r'\b{}\b'.format(search_word), all_messages, flags=re.IGNORECASE))

    return word_count
```

```python
search_word = "Bhai"
word_count = count_word_occurrences(df['message'], search_word)
print(f'The word "{search_word}" appears {word_count[search_word.lower()]} times in the chat.')
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
def calculate_daily_sentiment(df):
    sia = SentimentIntensityAnalyzer()
    df['SentimentScore'] = df['message'].apply(lambda x: sia.polarity_scores(x)['compound'])
    daily_sentiment = df.groupby(df['per_date'])['SentimentScore'].mean().reset_index()
    return daily_sentiment
```

```python
daily_sentiment = calculate_daily_sentiment(df[['per_date','message']])
```

```python
import matplotlib.pyplot as plt
from collections import Counter
from wordcloud import STOPWORDS
from googletrans import Translator

# Sample Hindi text
hindi_text = "рдирдорд╕реНрддреЗ рдЬрдЧрд╣ рд╡рд┐рд╢реЗрд╖рдЬреНрдЮ рд╡рд┐рд╢реНрд▓реЗрд╖рдг рдкрд░рд┐рд╕рд░ рд╡рд┐рдХрд╕рд┐рдд рд╡рд┐рд╢реЗрд╖рдЬреНрдЮ рдкреНрд░рдХреНрд░рд┐рдпрд╛ рдЕрдиреБрд╕рдВрдзрд╛рди рд╕рдВрдЧреНрд░рд╣рд┐рдд"

# Translate Hindi text to English using GoogleTrans
translator = Translator()
english_text = translator.translate(hindi_text, src='hi', dest='en').text

# Remove common English stopwords
stopwords = set(STOPWORDS)

# Tokenize the text into words
words = english_text.split()

# Remove stopwords and create a Counter object
word_counts = Counter(word for word in words if word not in stopwords)

# Get the top N most common words
top_n = 10  # You can adjust this number based on your preference
common_words = [word for word, _ in word_counts.most_common(top_n)]

# Plot the most common words using matplotlib with English font
plt.figure(figsize=(10, 5))
plt.bar(common_words, [word_counts[word] for word in common_words], color='skyblue')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title(f'Most Common {top_n} Words in Hindi Text')
plt.xticks(rotation=45, ha='right')
plt.show()
```
